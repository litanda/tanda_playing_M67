````
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
  plt.ylim([0,.1])
  plt.legend()

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
  plt.ylim([0,.2])
  plt.legend()
  plt.show()
````

````
dataset_path = glob.glob(rootdr + 'simple_grid_v3_sub1_100/*.csv')
dataset_path
````

````
#import data
initial_mass = []
star_age = []
initial_yinit = []
initial_feh = []
initial_mlt = []
initial_fov = []
initial_dif = []
effective_T = []
luminosity = []
log_g = []
nu_0_10 = []
nu_0_11 = []
nu_0_12 = []
nu_0_13 = []
for track in dataset_path:
    if os.path.isfile(track):
        print(track)
        raw_dataset = pd.read_csv(track)
        initial_mass =  np.concatenate([initial_mass, raw_dataset['initial_mass']])
        star_age = np.concatenate([star_age, raw_dataset['star_age']])
        initial_yinit = np.concatenate([initial_feh, raw_dataset['initial_Yinit']])
        initial_feh = np.concatenate([initial_feh, raw_dataset['initial_feh']])
        initial_mlt = np.concatenate([initial_mlt, raw_dataset['initial_MLT']])
        initial_fov =  np.concatenate([initial_fov, raw_dataset['initial_fov']])
        initial_dif =  np.concatenate([initial_dif, raw_dataset['initial_dif']])
        effective_T = np.concatenate([effective_T, raw_dataset['effective_T']])
        log_g = np.concatenate([log_g, raw_dataset['log_g']])
        luminosity = np.concatenate([luminosity, raw_dataset['luminosity']]) 
        nu_0_10 = np.concatenate([nu_0_10, raw_dataset['nu_0_10']]) 
        nu_0_11 = np.concatenate([nu_0_11, raw_dataset['nu_0_11']]) 
        nu_0_12 = np.concatenate([nu_0_12, raw_dataset['nu_0_12']]) 
        nu_0_13 = np.concatenate([nu_0_13, raw_dataset['nu_0_13']]) 
````

````
nuse = np.where((star_age >= 0.01) & (nu_0_10 > 0.00) & (nu_0_11 > 0.00) & (nu_0_12 > 0.00) & (nu_0_13 > 0.00))
    
print((initial_dif[nuse])
     )
````

````
bs = {'initial_mass':initial_mass[nuse],
      'star_age':np.log10(star_age[nuse]),
      'initial_yinit':initial_yinit[nuse],
      'initial_feh': initial_feh[nuse],
      'initial_mlt': initial_mlt[nuse],
      'initial_fov': initial_fov[nuse],
      'initial_dif': initial_dif[nuse],
      'effective_T':np.log10(effective_T[nuse]),
      'luminosity':np.log10(luminosity[nuse]),
      'log_g': log_g[nuse],
      'nu_0_10':np.log10(nu_0_10[nuse]),
      'nu_0_11':np.log10(nu_0_11[nuse]),
      'nu_0_12':np.log10(nu_0_12[nuse]),
      'nu_0_13':np.log10(nu_0_13[nuse])
     }

dataset = pd.DataFrame(bs, columns= ['initial_mass', 'star_age','initial_feh','initial_mlt','initial_fov','initial_dif','effective_T','log_g','luminosity','nu_0_10', 'nu_0_11','nu_0_12','nu_0_13'])
dataset.isna().sum()
dataset = dataset.dropna()
dataset = dataset.sample(frac = 1.0, random_state=1)
dataset.tail()
````

````
sns.pairplot(dataset[['effective_T','log_g','nu_0_10', 'nu_0_13']])
````

````

ts = {'initial_mass':initial_mass,
      'star_age':np.log10(star_age),
      'initial_yinit':initial_yinit,
      'initial_feh': initial_feh,
      'initial_mlt': initial_mlt,
      'initial_fov': initial_fov,
      'initial_dif': initial_dif,
      'effective_T':np.log10(effective_T),
      'luminosity':np.log10(luminosity),
      'log_g': log_g
     }


datatest = pd.DataFrame(ts,  columns= ['initial_mass', 'star_age','initial_feh','initial_mlt','initial_fov','initial_dif','effective_T','log_g','luminosity'])
datatest.isna().sum()
datatest = datatest.dropna()
datatest = datatest.sample(frac = 1.0, random_state=1)
datatest.tail()
````

# Build the model
Let's build our model. Here, we'll use a Sequential model with two densely connected hidden layers, and an output layer that returns a single, continuous value. The model building steps are wrapped in a function, build_model, since we'll create a second model, later on

````
train_dataset = dataset[['initial_mass', 'star_age','initial_feh','initial_mlt','initial_fov','initial_dif','effective_T','log_g','luminosity']]
train_labels = dataset[["nu_0_10","nu_0_11","nu_0_12","nu_0_13"]]
test_dataset = datatest[['initial_mass', 'star_age','initial_feh','initial_mlt','initial_fov','initial_dif','effective_T','log_g','luminosity']]
````

````
from keras import regularizers
````

````
model = tf.keras.Sequential()
model.add(layers.Dense(128, activation ='elu', input_shape=[len(train_dataset.keys())],kernel_regularizer=regularizers.l2(0.000001)))
model.add(layers.Dense(128, activation ='elu',kernel_regularizer=regularizers.l2(0.000001)))
model.add(layers.Dense(128, activation ='elu',kernel_regularizer=regularizers.l2(0.000001)))
model.add(layers.Dense(128, activation ='elu',kernel_regularizer=regularizers.l2(0.000001)))
model.add(layers.Dense(128, activation ='elu',kernel_regularizer=regularizers.l2(0.000001)))
model.add(layers.Dense(128, activation ='elu',kernel_regularizer=regularizers.l2(0.000001)))
model.add(layers.Dense(4, activation ='linear'))

#optimizer = tf.keras.optimizers.RMSprop(0.001)
#optimizer = tf.keras.optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
optimizer = tf.keras.optimizers.Nadam(learning_rate=0.001, beta_1=0.9995, beta_2=0.9999)
          
model.compile(loss='mae',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
#Use the .summary method to print a simple description of the model
#train model

# Display training progress by printing a single dot for each completed epoch
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')

#EPOCHS = 1000
# The patience parameter is the amount of epochs to check for improvement
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0.001, patience = 10000)

EPOCHS = 100000

history = model.fit(train_dataset, train_labels, epochs=EPOCHS, 
                    batch_size= int(0.6*len(train_dataset)),
                    validation_split = 0.4, verbose=0, callbacks=[early_stop, PrintDot()])#callbacks = [PrintDot()])

model.save(rootdr + '20200125_frequencies.h5')

hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
#hist.tail()

````

````

loss, mae, mse = model.evaluate(train_dataset, train_labels, verbose=2)

print("Testing set Mean Abs Error: {:5.2f} ".format(mae))
````

````
train_predictions = model.predict(train_dataset) #.flatten()

fig = plt.figure()
plt.scatter(train_dataset['effective_T'], train_labels['nu_0_10'], color = 'black', s = 0.5)
plt.scatter(train_dataset['effective_T'],train_predictions[:,0], color = 'red', s = 0.5)
plt.scatter(train_dataset['effective_T'], train_labels['nu_0_13'], color = 'black', s = 0.5)
plt.scatter(train_dataset['effective_T'],train_predictions[:,3], color = 'blue', s = 0.5)
plt.xlabel('logT')
plt.ylabel('log(nu_0_1)')
plt.show()
fig.savefig(rootdr + '20200127_nu_0_1.png')
````

````
error_T = 10**train_predictions[:,0] - 10**train_labels['nu_0_10']
plt.hist(error_T, bins = 200,range = [-0.02,0.02])
plt.xlabel("Prediction Error [nu_0_1] (microHz)")
_ = plt.ylabel("Count")
fig.savefig(rootdr + '20200127_train_prediction_nu_0_1.png')
````

````
test_predictions = model.predict(test_dataset) #.flatten()

fig = plt.figure()
plt.scatter(test_dataset['effective_T'], test_predictions[:,0], color = 'red', s = 0.5)
plt.xlabel('logT')
plt.ylabel('log(nu_0_10)')
plt.show()
fig.savefig(rootdr + 'nu_0_10_inter_grid.png')
````
